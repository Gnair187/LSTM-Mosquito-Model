```python
# Script to extract climate data for the Greater Toronto Area. https://climate.weather.gc.ca/historical_data/search_historic_data_e.html
# Station names of climate gauges. These stations were selected as they were near station files. 
# Closest stations were selected 
# stationnames=['4525', '27604', '10911', '31010', '45407', '27545', '4923', '7868', '27529', '4991', '31688', 
#                '26953', '53678', '4841', '4996', '51459', '5097', '51938', '52704', '52678', '48649']
```

```python
import os
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np
from mpl_toolkits.basemap import Basemap

folder_path = r"Path\To\File\Climate Records Dataset" # This file contains compiled climate data from the GTA region, from 2000 to 2017
shapefile_path = r"Path\To\File\GTA.shp"

gdf_shape = gpd.read_file(shapefile_path)
dfs = []

for file in os.listdir(folder_path):
    if file.endswith('.csv'):
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        
        # Use the long lat to see if the data point falls in the shp.
        gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.x, df.y))
        gdf.crs = "EPSG:4326"
        points_in_shape = gpd.sjoin(gdf, gdf_shape, how="inner", op="intersects")
        dfs.append(points_in_shape)

result = pd.concat(dfs, ignore_index=True)
```

```python
# Drop unnecessary columns
result=result.drop(columns=['CLIMATE_IDENTIFIER', 'LOCAL_YEAR',
       'LOCAL_MONTH', 'LOCAL_DAY', 'MIN_TEMPERATURE',
       'MAX_TEMPERATURE', 'TOTAL_SNOW',
       'SNOW_ON_GROUND', 'DIRECTION_MAX_GUST', 'SPEED_MAX_GUST',
       'COOLING_DEGREE_DAYS', 'HEATING_DEGREE_DAYS', 'MIN_REL_HUMIDITY',
       'MIN_REL_HUMIDITY_FLAG', 'MAX_REL_HUMIDITY', 'MAX_REL_HUMIDITY_FLAG',
       'index_right', 'GEODB_OID', 'OBJECTID', 'OBJECT_ID', 'GUT_NUMBER',
       'SHAPE_AREA', 'SHAPE_LEN', 'DIRECTION_MAX_GUST_FLAG',
       'SPEED_MAX_GUST_FLAG', 'MEAN_TEMPERATURE_FLAG', 'MIN_TEMPERATURE_FLAG',
       'MAX_TEMPERATURE_FLAG', 'COOLING_DEGREE_DAYS_FLAG',
       'HEATING_DEGREE_DAYS_FLAG', 'STATION_NAME', 'ID', 'PROVINCE_CODE',
       'TOTAL_PRECIPITATION_FLAG', 'TOTAL_RAIN_FLAG', 'TOTAL_SNOW_FLAG',
       'SNOW_ON_GROUND_FLAG', 'DESCR', 'OFF_NAME', 'LOC_NAME'])
```

```python
resultfinal = result.groupby('LOCAL_DATE').mean().reset_index()
resultfinal['LOCAL_DATE'] = pd.to_datetime(resultfinal['LOCAL_DATE'])
resultfinal['Year'] = resultfinal['LOCAL_DATE'].dt.year
resultfinal['Day'] = resultfinal['LOCAL_DATE'].dt.day
resultfinal['Month'] = resultfinal['LOCAL_DATE'].dt.month

new_column_names = {
    'LOCAL_DATE': 'Date',
    'x': 'Longitude (x)',
    'y': 'Latitude (y)',
    'MEAN_TEMPERATURE': 'Mean Temp (°C)',
    'TOTAL_RAIN': 'Total Rain (mm)',
    'TOTAL_PRECIPITATION': 'Total Precip (mm)',
    'Year': 'Year',
    'Month': 'Month',
    'Day': 'Day'
}

resultfinal.rename(columns=new_column_names, inplace=True)
```

```python
import copy
compiledobserved=copy.deepcopy(resultfinal)
compiledobserved['Date']=pd.to_datetime(compiledobserved[["Year", "Month", "Day"]])

# This script gives climate values from 2000 to 2017 
compiledobserved.to_excel(r'Path\To\File\ClimateDataNew2023.xlsx')
```

```python
# Group by week and month, calculate mean values, and reset index
CMIP_and_Observed_Weekly = compiledobserved.groupby(pd.Grouper(key='Date', freq='W')).mean()
CMIP_and_Observed_Weekly = CMIP_and_Observed_Weekly.reset_index()

CMIP_and_Observed_Monthly = compiledobserved.groupby(pd.Grouper(key='Date', freq='M')).mean()
CMIP_and_Observed_Monthly = CMIP_and_Observed_Monthly.reset_index()
```

```python
TestingforTemp=compiledobserved.groupby(pd.Grouper(key='Date', freq='M')).mean() 
TestingforTemp=TestingforTemp.reset_index()

TestingforTemp['Date']=pd.to_datetime(TestingforTemp['Date'])
TestingforTemp['Year']=TestingforTemp['Date'].dt.year
TestingforTemp['Month']=TestingforTemp['Date'].dt.month
TestingforTemp['Day']=TestingforTemp['Date'].dt.day
TempTestData=TestingforTemp.groupby(['Month']).mean()
```

```python
TestingforPrecip=compiledobserved.groupby(pd.Grouper(key='Date', freq='M')).sum() 
TestingforPrecip=TestingforPrecip.reset_index()

TestingforPrecip['Date']=pd.to_datetime(TestingforPrecip['Date'])
TestingforPrecip['Year']=TestingforPrecip['Date'].dt.year
TestingforPrecip['Month']=TestingforPrecip['Date'].dt.month
TestingforPrecip['Day']=TestingforPrecip['Date'].dt.day

PrecipTestData=TestingforPrecip.groupby(['Month']).mean()
```
